```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = np.zeros((600,600,3), dtype=np.uint8)

white=[255,255,255]
saffron=[255,153,51]
green=[19,136,8]

img[:200,:,:]=saffron
img[200:400,:,:]=white
img[400:,:,:]=green
```

```python
spokes=24
radius=100
theta=np.linspace(0,2*np.pi,spokes,endpoint=False)
spoke_length=radius*1

for angle in theta:
    x=int(300+spoke_length*np.cos(angle))
    y=int(300+spoke_length*np.sin(angle))
    cv2.line(img,(300,300),(x,y),(0,0,128), thickness=5)
```

```python
cv2.circle(img, (300,300),100,(0,0,128),thickness=5)
```

```python
plt.imshow(img)
```
